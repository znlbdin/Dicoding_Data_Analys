# Cleaning Data

```python
#duplicated data from customers table
#method drop_duplicates() -> dropping the duplicated data from table
customers_df.drop_duplicates(inplace=True)
```

```python
#cheking the result after dropping the duplicated data
print('Jumlah duplikasi data Customers: ',customers_df.duplicated().sum())
```

```python
#missing value data from customers table
#check the missing value in table customers
customers_df[customers_df.gender.isna()]
```

Because of the data that is missing value containing to much of important data, the way to solve this is to impute dominan data

```python
#cheking the dominan data that is NaN
#method value_counts() -> identified the dominan data
customers_df.gender.value_counts()
```

Dapat diketahui bahwa nilai yang paling dominan dalam kolom gender ialah “Prefer not to say”. Nilai inilah yang selanjutnya akan kita gunakan sebagai pengganti missing value

```python
#method fillna() -> replacing the NaN value to the dominan value from the code above
customers_df.fillna(value='Prefer not to say',inplace=True)
```

```python
#checking the missing value after filling the missing
customers_df.isna().sum()
```

```python
#checking the inacuarte value of customers table on age column
customers_df[customers_df.age == customers_df.age.max()]
```

Berdasarkan data tersebut, kita bisa berasumsi bahwa inaccurate value tersebut terjadi karena human error sehingga kelebihan memasukkan nilai nol

```python
#replace the value of age to 70 because of human error filling
customers_df.age.replace(customers_df.age.max(),70,inplace=True)
```

```python
#checking the maximal age again
customers_df[customers_df.age == customers_df.age.max()]
```

```python
#ups there is error human filling again for age columns
#replace again with metode replace()
customers_df.age.replace(customers_df.age.max(),50,inplace=True)
```

```python
#checking the maximal age
customers_df[customers_df.age == customers_df.age.max()]
```

```python
customers_df.describe()
```

```python
#changing the data type of orders table
#method to_datetime() -> to chane the data type to datetime format
datetime_columns = ['order_date','delivery_date']

for columns in datetime_columns:
  orders_df[columns] = pd.to_datetime(orders_df[columns])
```

```python
#checking the orders table info
orders_df.info()
```

```python
#dropping duplicated data from products table
#method drop_duplicates() -> removing duplicated data from products table
products_df.drop_duplicates(inplace=True)
```

```python
print('Jumlah duplikasi data Products: ',products_df.duplicated().sum())
```

```python
#checking the missing value data from sales table
sales_df[sales_df.total_price.isna()]
```

Berdasarkan tampilan data tersebut, kita menemukan bahwa nilai total_price merupakan hasil perkalian antara price_per_unit dan quantity. Kita dapat menggunakan pola ini untuk menangani missing value pada kolom total_price

```python
#making measure for total_price that is total_price is price_per_unit * quantity
sales_df['total_price'] = sales_df['price_per_unit'] * sales_df['quantity']
```
